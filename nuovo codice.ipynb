```python
# Impostiamo il grafico
plt.figure(figsize=(8, 5))

# Aggiungiamo 'hue' per il groupby (es. 'gender')
ax = sns.countplot(data=df, x='diabetes', hue='gender', palette=['blue', 'hotpink'])

# Aggiungiamo i conteggi sopra ogni barra
for container in ax.containers:
    ax.bar_label(container)

plt.title('Distribuzione Diabete raggruppata per Genere')
plt.xlabel('Diabete')
plt.ylabel('Numero di pazienti')
plt.legend(title='Genere')
plt.show()
```

```python
#Plot histograms for varibiability in other columns/variables

sns.set_style("darkgrid")

numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Usiamo lo Scatter Plot che Ã¨ perfetto per Glucosio vs Insulina
# Usiamo 'hue' per colorare i punti in base a chi ha il diabete (0 o 1)
sns.scatterplot(x="glucose", y="insulin", data=df, hue="diabetes", palette='viridis', alpha=0.5)

plt.title('Relazione tra Glucosio e Insulina (Divisa per Diabete)')
plt.xlabel('Glucosio')
plt.ylabel('Insulina')
plt.legend(title='Diabete (1=SÃ¬, 0=No)')
plt.show()
```

```python
plt.figure(figsize=(5,5))

sns.violinplot(
    x=df["diabetes"],
    y=df["insulin"],
    data=df,
    palette="viridis", 
    alpha=0.7
)

plt.title('Violin Plot for Diabetes and Insulin')
plt.xlabel('Diabetes')
plt.ylabel('Insulin')
plt.show()

```

```python
df['glucose_group'] = pd.cut(df['glucose'],
                             bins=[70, 90, 110, 130, 150, 170, 200],
                             labels=['70-90','91-110','111-130','131-150','151-170','171-200'])

palette = {
    '70-90': 'lightgrey',
    '91-110': 'lightblue',
    '111-130': 'lightgreen',
    '131-150': 'gold',
    '151-170': 'lightskyblue',
    '171-200': 'lightblack'
}

plt.figure(figsize=(14, 8))
sns.violinplot(x='diabetes', y='glucose', hue='glucose_group', data=df, palette=palette, alpha=0.7)
plt.title('Violin Plot Diabetes vs Glucose')
plt.xlabel('Diabetes')
plt.ylabel('Glucose')
plt.show()

# "In questo grafico abbiamo diviso i pazienti in base alla loro diagnosi (asse X) e 
# abbiamo colorato i dati in base a sei fasce di glicemia. Nonostante la distinzione cromatica, 
# osserviamo che la densitÃ  di pazienti con glucosio alto (fascia 171-200, in giallo)
#  Ã¨ simile sia nel gruppo dei sani che in quello dei diabetici.
#  Questo conferma visivamente che, in questo campione, 
# il solo livello di glucosio non Ã¨ un fattore discriminante univoco per la diagnosi."
```

```python
cols_to_compare = ['age', 'bmi', 'glucose', 'blood_pressure', 'cholesterol', 'insulin', 'diabetes']
correlation_matrix = df[cols_to_compare].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlazione tra Parametri Clinici')
plt.show()
```
