```python
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], kde=True, color='orange')
plt.title('Distribuzione dell\'Et nel Campione')
plt.show()
```

```python
# Confronto Diet Score tra chi ha il diabete e chi no
plt.figure(figsize=(10, 6))
sns.boxplot(x='diabetes', y='diet_score', data=df, palette='magma')
plt.title('Relazione tra Qualit della Dieta e Diabete')
plt.xlabel('Diabete (0: No, 1: S)')
plt.ylabel('Diet Score')
plt.show()
```

```python
#Plot histograms for varibiability in other columns/variables
#droppata la id perch inutile alla data visualization 
exclude_cols = ['id', 'family_history', 'alcohol_intake', 'smoking', 'diabetes']
filtered_columns = [col for col in df.select_dtypes(include=["int64", "float64"]).columns if col not in exclude_cols]

n_cols = 2
n_rows = (len(filtered_columns) + 1) // n_cols

plt.figure(figsize=(14, n_rows * 4))
for idx, feature in enumerate(filtered_columns, 1):
    plt.subplot(n_rows, n_cols, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Usiamo lo Scatter Plot che  perfetto per Glucosio vs Insulina
# Usiamo 'hue' per colorare i punti in base a chi ha il diabete (0 o 1)
sns.scatterplot(x="glucose", y="insulin", data=df, hue="diabetes", palette='viridis', alpha=0.5)

plt.title('Relazione tra Glucosio e Insulina (Divisa per Diabete)')
plt.xlabel('Glucosio')
plt.ylabel('Insulina')
plt.legend(title='Diabete (1=S, 0=No)')
plt.show()
```

```python
plt.figure(figsize=(5,5))

sns.violinplot(
    x=df["diabetes"],
    y=df["insulin"],
    data=df,
    palette="viridis", 
    alpha=0.7
)

plt.title('Violin Plot for Diabetes and Insulin')
plt.xlabel('Diabetes')
plt.ylabel('Insulin')
plt.show()

```

```python
df['glucose_group'] = pd.cut(df['glucose'],
                             bins=[70, 90, 110, 130, 150, 170, 200],
                             labels=['70-90','91-110','111-130','131-150','151-170','171-200'])

palette = {
    '70-90': 'lightgrey',
    '91-110': 'lightblue',
    '111-130': 'lightgreen',
    '131-150': 'gold',
    '151-170': 'lightskyblue',
    '171-200': 'lightblack'
}

plt.figure(figsize=(14, 8))
sns.violinplot(x='diabetes', y='glucose', hue='glucose_group', data=df, palette=palette, alpha=0.7)
plt.title('Violin Plot Diabetes vs Glucose')
plt.xlabel('Diabetes')
plt.ylabel('Glucose')
plt.show()

# "In questo grafico abbiamo diviso i pazienti in base alla loro diagnosi (asse X) e 
# abbiamo colorato i dati in base a sei fasce di glicemia. Nonostante la distinzione cromatica, 
# osserviamo che la densit di pazienti con glucosio alto (fascia 171-200, in giallo)
#   simile sia nel gruppo dei sani che in quello dei diabetici.
#  Questo conferma visivamente che, in questo campione, 
# il solo livello di glucosio non  un fattore discriminante univoco per la diagnosi."
```
